```python
import pandas as pd
from sklearn.preprocessing import StandardScaler
import numpy as np

# Columns to scale
numerical_columns = ['dur', 'tot_pkts', 'tot_bytes', 'src_bytes']

# Initialize the scaler
scaler = StandardScaler()

# Function to process data in batches and save each batch to disk
def scale_in_batches(df, batch_size=100000):
    n_batches = int(np.ceil(len(df) / batch_size))  # Calculate number of batches
    for i in range(n_batches):
        print(f"Processing batch {i+1}/{n_batches}...")
        
        # Process a batch
        batch = df.iloc[i * batch_size:(i + 1) * batch_size].copy()  # Copy a batch of data
        batch[numerical_columns] = scaler.fit_transform(batch[numerical_columns])  # Scale the batch
        
        # Save the batch to a CSV file to avoid memory overload
        batch_file = f"scaled_batch_{i+1}.csv"
        
        batch.to_csv(batch_file, index=False)
        print(f"Batch {i+1} saved to {batch_file}")
    
    print("All batches processed and saved to disk.")

# Apply batch scaling with smaller batch size
scale_in_batches(encoded_df)
```

```python
import pandas as pd
from sklearn.preprocessing import StandardScaler
import numpy as np

# Columns to scale
numerical_columns = ['dur', 'tot_pkts', 'tot_bytes', 'src_bytes']

# Function to process one column at a time in chunks and save to disk
def scale_column_in_chunks(df, column, chunk_size=25000):
    n_chunks = int(np.ceil(len(df) / chunk_size))  # Calculate number of chunks
    
    # Initialize the scaler
    scaler = StandardScaler()
    
    for i in range(n_chunks):
        print(f"Processing chunk {i+1}/{n_chunks} for column {column}...")
        
        # Process a chunk of the current column
        chunk = df.iloc[i * chunk_size:(i + 1) * chunk_size, :].copy()
        
        # Downcast to reduce memory usage
        chunk[column] = pd.to_numeric(chunk[column], downcast='float')
        
        # Scale the current column
        chunk[[column]] = scaler.fit_transform(chunk[[column]])
        
        # Save the chunk to a CSV file
        chunk_file = f"scaled_chunk_{column}_{i+1}.csv"
        chunk.to_csv(chunk_file, index=False)
        print(f"Chunk {i+1} for column {column} saved to {chunk_file}")
    
    print(f"All chunks for column {column} processed and saved to disk.")

# Scale each numerical column one at a time
for col in numerical_columns:
    scale_column_in_chunks(encoded_df, col)
```

```python
import pandas as pd
import os

# Specify the directory where the chunks are saved
chunk_dir = './'  # Adjust this if necessary

# Get all chunk files for each column
chunk_files = {
    'dur': [f for f in os.listdir(chunk_dir) if 'scaled_chunk_dur' in f],
    'tot_pkts': [f for f in os.listdir(chunk_dir) if 'scaled_chunk_tot_pkts' in f],
    'tot_bytes': [f for f in os.listdir(chunk_dir) if 'scaled_chunk_tot_bytes' in f],
    'src_bytes': [f for f in os.listdir(chunk_dir) if 'scaled_chunk_src_bytes' in f]
}

# Initialize an empty dataframe to combine the chunks
combined_df = None

# Combine chunks for each column, handling different chunk sizes
for col, files in chunk_files.items():
    for i, file in enumerate(sorted(files)):
        chunk_df = pd.read_csv(os.path.join(chunk_dir, file))[[col]]
        
        if combined_df is None:
            combined_df = pd.DataFrame(index=chunk_df.index)  # Initialize with index only

        # Handle different chunk sizes
        combined_df = combined_df.reindex(chunk_df.index)  # Ensure the index matches
        combined_df[col] = chunk_df[col].values  # Add each column to the combined dataframe

# Save the combined dataframe to a new CSV file
combined_df.to_csv('combined_scaled_data.csv', index=False)
print("All columns combined and saved to 'combined_scaled_data.csv'.")
```

```python
import pandas as pd
from sklearn.cluster import KMeans
import pickle

# Load the preprocessed data
data = pd.read_csv('preprocessed_data.csv')

# Initialize and train the K-Means model
kmeans = KMeans(n_clusters=3, random_state=42)
kmeans.fit(data)

# Save the K-Means model for future use
with open('kmeans_model.pkl', 'wb') as file:
    pickle.dump(kmeans, file)

# Add the cluster labels to the dataframe
data['Cluster'] = kmeans.labels_

# Save the clustered data
data.to_csv('kmeans_clustered_data.csv', index=False)
print("K-Means clustering complete. Clustered data saved to 'kmeans_clustered_data.csv'.")
```

```python
from sklearn.metrics import silhouette_score, davies_bouldin_score

# Load the clustered data
data = pd.read_csv('kmeans_clustered_data.csv')

# Silhouette Score (measures how similar each point is to its own cluster)
silhouette_avg = silhouette_score(data.drop('Cluster', axis=1), data['Cluster'])
print("Silhouette Score for K-Means:", silhouette_avg)

# Inertia (Sum of Squared Distances of samples to their closest cluster center)
inertia = kmeans.inertia_
print("K-Means Inertia:", inertia)

# Davies-Bouldin Index (lower is better)
db_score = davies_bouldin_score(data.drop('Cluster', axis=1), data['Cluster'])
print("Davies-Bouldin Score for K-Means:", db_score)

# Store the metrics for later comparison
metrics_kmeans = {
    'Silhouette Score': silhouette_avg,
    'Inertia': inertia,
    'Davies-Bouldin Score': db_score
}

# Print the stored metrics
print("K-Means Metrics:", metrics_kmeans)
```

```python
from sklearn.cluster import DBSCAN
from sklearn.metrics import silhouette_score, davies_bouldin_score
import pandas as pd
import numpy as np

# Load the preprocessed data
data = pd.read_csv('preprocessed_data.csv')

# Range of values to test for 'eps' and 'min_samples'
eps_values = np.arange(0.5, 2.0, 0.1)  # Testing 'eps' values from 0.5 to 2.0 with a step of 0.1
min_samples_values = range(2, 10)  # Testing 'min_samples' from 2 to 9

# Variables to track the best scores and parameters
best_eps = None
best_min_samples = None
best_silhouette = -1  # Initialize with the worst score
best_db_score = None

# Loop over the range of 'eps' and 'min_samples'
for eps in eps_values:
    for min_samples in min_samples_values:
        # Train DBSCAN with the current combination of eps and min_samples
        dbscan = DBSCAN(eps=eps, min_samples=min_samples)
        labels = dbscan.fit_predict(data)
        
        # Check if there are at least 2 clusters (skip cases with fewer clusters)
        unique_labels = set(labels)
        if len(unique_labels) < 2 or -1 in unique_labels and len(unique_labels) == 2:
            print(f"Skipping eps={eps}, min_samples={min_samples} due to insufficient clusters.")
            continue
        
        # Filter out noise points (-1) for the metrics calculation
        filtered_data = data[labels != -1]
        filtered_labels = labels[labels != -1]
        
        # Calculate Silhouette Score and Davies-Bouldin Score
        silhouette_avg = silhouette_score(filtered_data, filtered_labels)
        db_score = davies_bouldin_score(filtered_data, filtered_labels)
        
        # Check if the current model is better than the previous best
        if silhouette_avg > best_silhouette:
            best_silhouette = silhouette_avg
            best_eps = eps
            best_min_samples = min_samples
            best_db_score = db_score
        
        # Print progress
        print(f"Tested eps={eps}, min_samples={min_samples}: Silhouette={silhouette_avg:.4f}, DB={db_score:.4f}")

# Output the best combination found
print("\nBest parameters found:")
print(f"Best eps: {best_eps}")
print(f"Best min_samples: {best_min_samples}")
print(f"Best Silhouette Score: {best_silhouette}")
print(f"Best Davies-Bouldin Score: {best_db_score}")

# Retrain DBSCAN with the best found parameters and save the model
best_dbscan = DBSCAN(eps=best_eps, min_samples=best_min_samples)
data['Cluster'] = best_dbscan.fit_predict(data)

# Save the retrained model
data.to_csv('dbscan_clustered_data_best.csv', index=False)
print(f"Best DBSCAN model retrained and saved to 'dbscan_clustered_data_best.csv'.")
```
